```
covid_data = pd.read_csv ("covid_19_india.csv")
```

```
covid_data
```

```
covid_data.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational','Deaths','Confirmed'],axis = 1,inplace = True)
```

```
covid_data
```

```
covid_19_data = covid_data.pivot_table('Cured',['Date'], 'State/UnionTerritory')
```

```
covid_19_data
```

```
covid_19_data.fillna(0, inplace=True)
```

```
covid_19_data
```

```
covid_19_data.sort_values(list(covid_19_data.columns),inplace=True)
```

```
covid_19_data
```

```
covid_19_data = covid_19_data.iloc[62:,]
```

```
covid_19_data
```

```
import bar_chart_race as bcr
```

```
bcr.bar_chart_race(
    df=covid_19_data,
    filename=None,
    orientation='h',
    sort='desc',
    n_bars=10,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.90,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total recover: {v.nlargest(39).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    period_length=500,
    figsize=(5,3.3 ),
    dpi=144,
    cmap='G10',
    title='India\'s COVID-19 recovery by State' ,
    title_size='',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Helvetica', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=True)
```
